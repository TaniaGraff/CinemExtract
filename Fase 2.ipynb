## Fase 1: Extracción de Datos de API de MoviesDataset

```python
import requests

def peliculas_uno(genero, anyo):

	url = "https://moviesdatabase.p.rapidapi.com/titles"
 #diccionario donde se guarda cada pagina
	paginas = {}
	
	#recorro de la pagina 1 a la 10
	for p in range (1,10):
     
		querystring =  {"genre":{genero},"year":{anyo},"page":{p},"limit":"50"}


		headers = {
			"X-RapidAPI-Key": "96d0702058msh370b85c6df311dcp1b348cjsn7bebff70ced1",
			"X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"
		}

		response = requests.get(url, headers=headers, params=querystring)
	#lo guardo dentro de un diccionario nuevo dentro del diccionario
		paginas[p] = response.json()
		 
 #devuelvo el diccionario con las paginas dentro
	return paginas
```

### Convertir a lista de tuplas

```python
def tuplas_en_lista(respuestas):

    lista_tuplas = []

    for p in respuestas.values():
        #genero 
        cadena = p["next"]
        genero = cadena.split('=')[-1]
        #iterar en results
        for result in p["results"]:
            peliculas = []
            if result["titleType"]["text"] != "Short" and result["titleType"]["text"] != "Movie":
              pass
            else:
              if result["releaseDate"] is not None :
              #tipo
                peliculas.append(result["titleType"]["text"])
              #nombre
                peliculas.append(result["titleText"]["text"])
              #año
                peliculas.append(result["releaseYear"]["year"])
              #mes
                peliculas.append(result["releaseDate"]["month"])
              #id
                peliculas.append(result["id"])
              #genero
                peliculas.append(genero)
              else:
                continue
   
              lista_tuplas.append(tuple(peliculas))
    return lista_tuplas
```

## Fase 2: Extracción de Detalles de Películas con Selenium

```python
# Funcion para sacar datos de peliculas en IMBD a partir de una lista de tuplas de peliculas
def datos_imdb(lista_movies):
    
    # Accedo a los id de las peliculas y los guardo en una lista para iterar por ellos
    lista_id_movies = [movie[-2] for movie in lista_movies]
    # Crear lista donde ir guardando las tuplas de cada pelicula
    lista_imbd = []

    # Inicializa el navegador Chrome
    driver = webdriver.Chrome()
    # Navega a la página web "https://www.imdb.com/"
    driver.get("https://www.imdb.com/")
    # Espera durante 3 segundos (para permitir que la página cargue completamente)
    sleep(2)
    # Acepta las cookies haciendo clic en el elemento con el XPath
    try:
        driver.find_element("xpath", "/html/body/div[2]/div/div/div[2]/div/button[2]").click()
    except:
        pass
    sleep(2)
    # Maximiza la ventana
    driver.maximize_window()
    sleep(2)
        
    # Itero por la lista de id de peliculas
    for id_movie in lista_id_movies:
        
        # Lista que recoge la información
        lista_movie = []
        
        # Intentar encontrar la pelicula
        try:
            # Introducir id de pelicula en el campo de búsqueda y presionar Enter
            driver.find_element("css selector", "#suggestion-search").send_keys(id_movie, Keys.ENTER)             

            sleep(2)
            # Accede a cada uno de los datos que quiero guardar en el diccionario:
            
        except:
            # Pruebo desde la pagina principal
            driver.get("https://www.imdb.com/")
            sleep(2)
            # Introducir id de pelicula en el campo de búsqueda y presionar Enter
            driver.find_element("css selector", "#suggestion-search").send_keys(id_movie, Keys.ENTER)             

            sleep(2)
            # Accede a cada uno de los datos que quiero guardar en el diccionario:
        
        # Puntuacion
        # Porcentaje
        try:
            puntuacion_string = driver.find_element("xpath", '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[2]/div/div[1]/a/span/div/div[2]/div[1]/span[1]').text
            puntuacion_string = puntuacion_string.replace(",",".")
            puntuacion = float(puntuacion_string)
            porcentaje = int(round(puntuacion*10,0))
        except:
            puntuacion = None
            porcentaje = None
        
        # Direccion
        try:
            direccion = driver.find_element("xpath", '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[1]/div/ul/li/a').text
        except:
            try:
                direccion = driver.find_element("xpath", '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div/div[1]/section/div[2]/div/ul/li[1]/div/ul/li/a').text
            except:
                direccion = None
        
        # Guion
        try:
            # Encontrar la sección de guionistas
            guion_section = driver.find_element("xpath", '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[2]/div/ul')
            # Obtener los nombres mediante la etiqueta <a>
            guion = []
            for link in guion_section.find_elements(By.TAG_NAME, 'a'):
                guion.append(link.text)
        except:
            try:
                guion_section = driver.find_element("xpath", '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div/div[1]/section/div[2]/div/ul/li[2]/div/ul')
                guion = []
                for link in guion_section.find_elements(By.TAG_NAME, 'a'):
                    guion.append(link.text)
            except:
                guion = None
        
        # Argumento
        try:
            argumento = driver.find_element("xpath", '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/p/span[3]').text
            # Si está vacío devuelve None
            if argumento == '':
                argumento = None
        except:
            # Si no ha encontrado el anterior path, pruebo con otro
            try:
                argumento = driver.find_element("xpath", '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div/div[1]/section/p/span[3]').text
                if argumento == '':
                    argumento = None
            except:
                argumento = None
        
        # Duracion
        try:
            duracion_str = driver.find_element("xpath", '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]').text
            # Compruebo si el xpath me lleva a los minutos
            if 'min' in duracion_str and 'h' in duracion_str:
                # Usamos expresiones regulares para extraer las horas y los minutos y lo convierto a minutos totales
                horas = re.findall(r'(\d+)h', duracion_str)
                minutos = re.findall(r'(\d+)min', duracion_str)
                duracion = int(horas[0]) * 60 + int(minutos[0])
            elif 'min' in duracion_str and 'h' not in duracion_str:
                minutos = re.findall(r'(\d+)min', duracion_str)
                duracion = int(minutos[0])
            elif 'h' in duracion_str and 'min' not in duracion_str:
                horas = re.findall(r'(\d+)h', duracion_str)
                duracion = int(horas[0]) * 60
            else:
                # Si no, uso otro xpath
                duracion_str = driver.find_element("xpath", '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[3]').text
                if 'min' in duracion_str and 'h' in duracion_str:
                    horas = re.findall(r'(\d+)h', duracion_str)
                    minutos = re.findall(r'(\d+)min', duracion_str)
                    duracion = int(horas[0]) * 60 + int(minutos[0])
                elif 'min' in duracion_str and 'h' not in duracion_str:
                    minutos = re.findall(r'(\d+)min', duracion_str)
                    duracion = int(minutos[0])
                elif 'h' in duracion_str and 'min' not in duracion_str:
                    horas = re.findall(r'(\d+)h', duracion_str)
                    duracion = int(horas[0]) * 60
                else:
                    pass
        except:
            duracion = None
        
        # Genero  
        try:
            # Encontrar la sección de genero
            genero_section = driver.find_element("xpath", '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]')
            # Obtener los generos mediante la etiqueta <a>
            genero = []
            for link in genero_section.find_elements(By.TAG_NAME, 'a'):
                # Si la etiqueta es Corto la ignoro
                if link.text == "Corto":
                    pass
                else:
                    genero.append(link.text)
        except:
            try:
                # Encontrar la sección de genero
                genero_section = driver.find_element("xpath", '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div/div[1]/section/div[1]/div[2]')
                # Obtener los generos mediante la etiqueta <a>
                genero = []
                for link in genero_section.find_elements(By.TAG_NAME, 'a'):
                    if link.text == "Corto":
                        pass
                    else:
                        genero.append(link.text)
            except:
                genero = None
        
        # Nombre 
        try:
            nombre_pelicula = driver.find_element("xpath", '/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/h1/span').text
        except:
            nombre_pelicula = None
            
        # Añado informacion en lista
        lista_movie.append(id_movie)
        lista_movie.append(puntuacion)
        lista_movie.append(porcentaje)
        lista_movie.append(direccion)
        lista_movie.append(guion)
        lista_movie.append(argumento)
        lista_movie.append(genero)
        lista_movie.append(duracion)
        lista_movie.append(nombre_pelicula)
        
        # Convierto la lista en tupla y la añado a la lista de tuplas
        lista_imbd.append(tuple(lista_movie))
            
    # Cierra el navegador
    driver.close()
    
    return lista_imbd
```
