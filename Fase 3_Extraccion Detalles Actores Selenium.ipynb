```python
def tuplas_en_lista(respuestas):

    lista_tuplas = []

    for p in respuestas.values():
        #genero 
        cadena = p["next"]
        genero = cadena.split('=')[-1]
        #iterar en results
        for result in p["results"]:
            peliculas = []
            if result["titleType"]["text"] != "Short" and result["titleType"]["text"] != "Movie":
              pass
            else:
              if result["releaseDate"] is not None :
              #tipo
                peliculas.append(result["titleType"]["text"])
              #nombre
                peliculas.append(result["titleText"]["text"])
              #año
                peliculas.append(result["releaseYear"]["year"])
              #mes
                peliculas.append(result["releaseDate"]["month"])
              #id
                peliculas.append(result["id"])
              #genero
                peliculas.append(genero)
              else:
                continue
   
              lista_tuplas.append(tuple(peliculas))
    return lista_tuplas
```

```python
# Funcion para sacar datos de los actores de las peliculas en IMBD a partir de una lista de tuplas de peliculas
def actores_imdb(lista_movies_ejemplo):
    
    # Inicializa el navegador Chrome
    driver = webdriver.Chrome()
    # Maximiza la ventana
    driver.maximize_window()
    # Navega a la página web "https://www.imdb.com/"
    driver.get("https://www.imdb.com/")
    # Espera durante 3 segundos (para permitir que la página cargue completamente)
    sleep(2)
    # Acepta las cookies haciendo clic en el elemento con el XPath
    driver.find_element("xpath", "/html/body/div[2]/div/div/div[2]/div/button[2]").click()
    
    #Inicializamos la lista donde guardaremos los detalles de los actores
    lista_actores = []

#Itero por la lista de peliculas por id
    for id in lista_movies_ejemplo:
    # Introducir id de pelicula en el campo de búsqueda y presionar Enter
        driver.find_element("css selector", "#suggestion-search").send_keys(id, Keys.ENTER)
        sleep (2)

        for i in range(1,11):
            try:
                actores = driver.find_element(By.CSS_SELECTOR, f"#__next > main > div > section.ipc-page-background.ipc-page-background--base.sc-c41b9732-0.NeSef > div > section > div > div.sc-978e9339-1.ihWZgK.ipc-page-grid__item.ipc-page-grid__item--span-2 > section.ipc-page-section.ipc-page-section--base.sc-bfec09a1-0.jgUBLM.title-cast.title-cast--movie.celwidget > div.ipc-shoveler.ipc-shoveler--base.ipc-shoveler--page0.title-cast__grid > div.ipc-sub-grid.ipc-sub-grid--page-span-2.ipc-sub-grid--wraps-at-above-l.ipc-shoveler__grid > div:nth-child({i}) > div.sc-bfec09a1-6.eTKJpq.title-cast-item__avatar > div > a > div")
                # Desplazar la página para hacer visible el elemento
                driver.execute_script("arguments[0].scrollIntoView(true);", actores)
                sleep(2)
                # Hacer clic en el elemento
                actores.click()
                sleep(2)
                #Extraemos 'Nombre'
                try:
                    nombre = driver.find_element ("xpath", '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div/h1/span').text
                    lista_actores.append(nombre)
                except:
                    try:
                        nombre = driver.find_element ("css selector", "#__next > main > div > section.ipc-page-background.ipc-page-background--base.sc-304f99f6-0.fSJiHR > section > div:nth-child(5) > section > section > div.sc-4e4cc5f9-3.dDRspk > div > h1 > span").text
                        lista_actores.append(nombre)
                    except(NoSuchElementException, JavascriptException, ElementClickInterceptedException, ElementNotInteractableException):
                        lista_actores.append('No info')
                        pass
                #Extraemos anyo de nacimiento
                try:
                    fecha_nac = driver.find_element ("xpath", '//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[6]/div[2]/ul/li[4]/div/ul/li[1]/a').text
                    patron = r'\b\d{4}\b'
                    anyo = re.findall(patron, fecha_nac)
                    lista_actores.extend(anyo)
                except:
                    try:
                        fecha_nac = driver.find_element ("xpath", '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/section/aside/div[1]/span[2]').text
                        patron = r'\b\d{4}\b'
                        anyo = re.findall(patron, fecha_nac)
                        lista_actores.extend(anyo)
                    except (NoSuchElementException, JavascriptException, ElementClickInterceptedException, ElementNotInteractableException):
                        lista_actores.append ('No info')
                        pass
                #Extraemos 'Conocido por'
                # Desplazar la página para hacer visible el elemento
                lista_conocido = []
                for c in range (1,4):
                    try: 
                        conocido = driver.find_element("xpath", f'//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/div[3]/section[1]/div[2]/div/div[2]/div[{c}]/div[2]/div[1]/a').text
                        lista_conocido.append(conocido)                      
                    except (NoSuchElementException, JavascriptException, ElementClickInterceptedException, ElementNotInteractableException):
                        lista_conocido.append ('No info')
                        pass
                lista_actores.append(lista_conocido)
                sleep(2)
                #Sacamos las profesiones     
                try:   
                    que_hace = driver.find_element("css selector", '#__next > main > div > section.ipc-page-background.ipc-page-background--base.sc-c41b9732-0.NeSef > section > div:nth-child(5) > section > section > div.sc-92625f35-3.frxYSZ > div > ul').text.replace("\n", ', ')
                    lista_que_hace = que_hace.split(', ')
                    lista_actores.append(lista_que_hace)
                except (NoSuchElementException, JavascriptException, ElementClickInterceptedException, ElementNotInteractableException):
                    lista_actores.append('No info')
                    pass
                #extraemos premios
                try:
                    premios_driver = driver.find_element("xpath", '//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[1]/div/ul/li/div/ul/li/span').text
                    patron_premios = r'(\d+) premios'
                    premios = re.findall(patron_premios, premios_driver)
                    lista_actores.append (premios)
                    driver.back()
                except (NoSuchElementException, JavascriptException, ElementClickInterceptedException, ElementNotInteractableException):
                    lista_actores.append(0)
                driver.back()
            except (NoSuchElementException, JavascriptException, ElementClickInterceptedException, ElementNotInteractableException):
                pass     
    driver.quit() 
    return lista_actores
```

```python
def transformar_tupla(lista_actores):
    # Transformar la lista en una lista de tuplas
    tupla_actores= []
    for i in range(0, len(lista_actores), 5):
        tupla = tuple(lista_actores[i:i+5])
        tupla_actores.append(tupla)
    return tupla_actores
```

```python
# Funcion limpiar los caracteres especiales de la lista de tuplas de peliculas
def limpiar_caracteres_especiales(lista_tuplas):

    lista_tuplas_limpia = []
    # Recorro la lista de tuplas
    for tupla in lista_tuplas:
        tupla_limpia = []
        # Recorro los elementos de la tupla y compruebo si es una string o una list
        for elemento in tupla:
            if isinstance(elemento, str):
                elemento_limpio = unidecode(elemento)
            elif isinstance(elemento, list):
                # Convertir la lista en una cadena separada por comas y limpiarla
                elemento_limpio = ', '.join([unidecode(s) for s in elemento if isinstance(s, str)])
            else:
                elemento_limpio = elemento
            tupla_limpia.append(elemento_limpio)
        lista_tuplas_limpia.append(tuple(tupla_limpia))
    return lista_tuplas_limpia
```

```python
#funciona para rellenar valores en la lista que no ha podido se extraidos
def replace_empty_lists(data):
    # Iterar sobre los elementos de la lista
    for i in range(len(data)):
        # Verificar si el elemento es una lista vacía
        if isinstance(data[i], list) and len(data[i]) == 0:
            # Reemplazar la lista vacía por '0'
            data[i] = 0
    return data
```

```python
# Función para agregar 'no info' después de un elemento específico
def add_no_info_after(data, target):
    # Iterar sobre los elementos de la lista
    for i in range(len(data)):
        # Verificar si el elemento actual es el elemento objetivo
        if data[i] == target:
            # Insertar 'no info' después del elemento objetivo
            data.insert(i + 1, 'no info')
            break  # Romper el bucle después de encontrar e insertar el valor
    return data
```
