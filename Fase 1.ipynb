```python
import requests

def peliculas_uno(genero, anyo):

	url = "https://moviesdatabase.p.rapidapi.com/titles"
 #diccionario donde se guarda cada pagina
	paginas = {}
	
	#recorro de la pagina 1 a la 10
	for p in range (1,10):
     
		querystring =  {"genre":{genero},"year":{anyo},"page":{p},"limit":"50"}


		headers = {
			"X-RapidAPI-Key": "96d0702058msh370b85c6df311dcp1b348cjsn7bebff70ced1",
			"X-RapidAPI-Host": "moviesdatabase.p.rapidapi.com"
		}

		response = requests.get(url, headers=headers, params=querystring)
	#lo guardo dentro de un diccionario nuevo dentro del diccionario
		paginas[p] = response.json()
		 
 #devuelvo el diccionario con las paginas dentro
	return paginas
```

## Convertir a lista de tuplas

```python
def tuplas_en_lista(respuestas):

    lista_tuplas = []

    for p in respuestas.values():
        #genero 
        cadena = p["next"]
        genero = cadena.split('=')[-1]
        #iterar en results
        for result in p["results"]:
            peliculas = []
            if result["titleType"]["text"] != "Short" and result["titleType"]["text"] != "Movie":
              pass
            else:
              if result["releaseDate"] is not None :
              #tipo
                peliculas.append(result["titleType"]["text"])
              #nombre
                peliculas.append(result["titleText"]["text"])
              #año
                peliculas.append(result["releaseYear"]["year"])
              #mes
                peliculas.append(result["releaseDate"]["month"])
              #id
                peliculas.append(result["id"])
              #genero
                peliculas.append(genero)
              else:
                continue
   
              lista_tuplas.append(tuple(peliculas))
    return lista_tuplas
```
